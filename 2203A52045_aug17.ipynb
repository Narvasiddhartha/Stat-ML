```python
import pandas as pd
a=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Salary_Data.csv')
print(a)
```

```python
x=a['YearsExperience']
y=a['Salary']
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
import numpy as np
x_transpose=x.transpose()
w=x_transpose*y
c=w*(x_transpose*x)**-1
print(c)
```

```python
yp=[]
for i in range(len(x)):
  z=w[i]*x[i]
  yp.append(z)
print(yp)
```

```python
error=[]
for i in range(len(x)):
  z=(y[i]-yp[i])**-2
  error.append(z)
mean_error=sum(error)/len(w)
print(mean_error)
```
