```python
x=df_train.drop(columns=['Crop_Damage','ID'],axis=1)
y=df_train['Crop_Damage']
```

```python
from imblearn.over_sampling import SMOTE    #using SMOTE Method
smt=SMOTE()
x,y=smt.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split

# Assuming train_data is your complete dataset
# Split the data into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)


# Print the shapes of the resulting sets
print("Training set shape:", x_train.shape, y_train.shape)
print("Testing set shape:", x_test.shape, y_test.shape)
```

```python
from sklearn.decomposition import PCA



# Initialize PCA for the desired number of components (let's say 10 components)
num_components = 5
pca = PCA(n_components=num_components)

# Fit and transform PCA on the training data
x_train_pca = pca.fit_transform(x_train)

# Transform testing and validation data based on the PCA from the training set
x_test_pca = pca.transform(x_test)

# Check the shape of transformed datasets
print("Transformed Training set shape:", x_train_pca.shape)
print("Transformed Testing set shape:", x_test_pca.shape)
```
